```python
avg_ld = np.mean(np.array(aligned), axis=0)
fig = plt.figure()
plt.plot(xaxis*1e6, avg_ld*PIXSIZE)
plt.xlabel('Position / um')
plt.ylabel('Atoms / #')
plt.title('Average Atom Profile over 25 runs')
```

```python
N_1d = avg_ld * PIXSIZE
```

```python
avg_norm = avg_ld / np.sum(avg_ld)
avg_fft = np.fft.fftshift(np.fft.fft(avg_norm, 1024))
```

```python
psd_avg = np.abs(avg_fft)**2
psd_norm = psd_avg / sum(psd_avg)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))

plt.semilogy(faxis[513:]*1e-6, psd_norm[513:])
#plt.xlim((0,40))
plt.xlabel('Wavenumber / um^-1')
plt.title('Power Spectrum of Averaged Atom Profiles')
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)

plt.plot(0.5 / faxis[513:] * 1e6, psd_norm[513:]*1e6) #factor of two from periodicity to feature size
plt.xlim((0,5))
plt.ylim((0, 1))
plt.xlabel('Resolution / um')
plt.ylabel('PSD / a.u.')
plt.title('Observed Peaks in Averaged Data PSD')
```

```python
tight_freq = 2*acm.pi*1000
long_freq = 2*acm.pi*10
trap_center = 140e-6

trap_field = (0.5*acm.M*long_freq**2 * (xaxis - trap_center)**2) / acm.MU_B
```

```python
min_ld = 2.0 / PIXSIZE
scaling = 150
recon_pot_all = -np.array([acm.HBAR * tight_freq * np.sqrt(1 + acm.A * scaling*nn) for nn in avg_ld])
print np.mean(recon_pot_all)
recon_pot = -np.array([acm.HBAR * tight_freq * np.sqrt(1 + acm.A * scaling*nn) if nn > min_ld else -np.mean(recon_pot_all) for nn in avg_ld])
#recon_pot -= np.mean(recon_pot)
recon_trap = -np.array([trap_field[nn] if dd > min_ld else 0 for nn, dd in enumerate(avg_ld)])
plt.plot(xaxis*1e6, recon_pot*1e9 / acm.MU_B + recon_trap*1e9)
```
