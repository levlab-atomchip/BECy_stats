```python
window = [350, 675, 50, 250]
dist.custom_fit_window = window
```

```python
expose_dict = defaultdict(list)
count_dict = defaultdict(list)

for this_file in dist.filelist:
    this_img = CI(this_file)
    this_img.truncate_image(*window)
    this_vva = this_img.get_variables_values()['ImagingVVA']
    
    this_counts = np.sum(np.int32(this_img.light_image_trunc) - np.int32(this_img.dark_image_trunc))

    expose_dict[this_vva].append(this_img.get_image_time())
    count_dict[this_vva].append(this_counts)
```

```python
from IPython.display import display, clear_output

old_filelist = []

expose_dict = defaultdict(list)
count_dict = defaultdict(list)

fig, ax = plt.subplots()

while(1):
    dist = CD(datadir, False)
    if len(dist.filelist) > len(old_filelist):
        
        clear_output(wait=True)
        
        new_filelist = dist.filelist
        
        new_files = set(new_filelist).difference(set(old_filelist))
        
        for this_file in new_files:
            this_img = CI(this_file)
            this_img.truncate_image(*window)
            this_vva = this_img.get_variables_values()['ImagingVVA']

            this_counts = np.sum(np.int32(this_img.light_image_trunc) - np.int32(this_img.dark_image_trunc))

            expose_dict[this_vva].append(this_img.get_image_time())
            count_dict[this_vva].append(this_counts)
        
        vvas = sorted(expose_dict.keys())
        for vva in vvas:
            #print image_time
            plt.plot(np.array(expose_dict[vva])*1e6, count_dict[vva], 'o', markersize=15, label='VVA %2.1f'%(vva))
            slope, intercept, r_value, _, std_err = \
                    stats.linregress(np.array(expose_dict[vva]), np.array(count_dict[vva]))
            exes = np.linspace(0, 100)*1e-6
            plt.plot(exes*1e6, (intercept + slope *exes), color='black' )
        plt.legend(loc='lower right')
        plt.ylim(0, )
        plt.xlim(0,)
        
        plt.show()
        
        
        old_filelist = new_filelist
    time.sleep(15)
        

#print intercept
```
