```python
sat_dict = {}
od_dict = {}
it_dict = {}

for this_file in dist.filelist:
    this_img = CI(this_file)
    if this_img.curr_cont_par >= 1.55:
        this_img.truncate_image(*window)
        this_time = this_img.get_image_time()
        if this_time not in sat_dict:
            sat_dict[this_time] = []
        if this_time not in od_dict:
            od_dict[this_time] = []
        if this_time not in it_dict:
            it_dict[this_time] = []
        sat_dict[this_time].append(this_img.saturation())
        od_dict[this_time].append(this_img.optdens_number())
        it_dict[this_time].append(this_img.int_term_number())
```

```python
image_times = sorted(sat_dict.keys())
for image_time in image_times:
    sort_sats, sort_ods = zip(*sorted(zip(sat_dict[image_time], od_dict[image_time])))
    plt.plot(sort_sats, sort_ods, 'o', markersize=10, label='%2.0f us'%(image_time*1e6))
plt.legend(loc='upper right')
plt.title('Optical Density Number')
plt.xlabel('Saturation')
plt.ylabel('Atom Number')

plt.figure()

for image_time in image_times:
    sort_sats, sort_its = zip(*sorted(zip(sat_dict[image_time], it_dict[image_time])))
    plt.plot(sort_sats, sort_its, 'o', markersize=10, label='%2.0f us'%(image_time*1e6))
plt.legend(loc='upper right')
plt.title('Intensity Correction Number')
plt.xlabel('Saturation')
plt.ylabel('Atom Number')
```

```python
def get_optimal_alpha(ods, its):
    ALPHAS = np.linspace(0, 2)
    num_spread = []
    
    for ALPHA in ALPHAS:
        int_corr_num = ALPHA*np.array(ods) + np.array(its)
        num_spread.append(np.std(int_corr_num))
    return ALPHAS[np.argmin(num_spread)]
```

```python
opt_alphas = []
opt_nums = []
for image_time in image_times:
    this_alpha = get_optimal_alpha(od_dict[image_time], it_dict[image_time])
    opt_alphas.append(this_alpha)
    this_num = np.mean(this_alpha * np.array(od_dict[image_time]) + np.array(it_dict[image_time]))
    opt_nums.append(this_num)
plt.plot(np.array(image_times)*1e6, opt_alphas, 'o', markersize=20)
plt.xlim(0, 100)
plt.figure()
plt.plot(np.array(image_times)*1e6, opt_nums, 'o', markersize=20)
```

```python
def get_optimal_chi(ods, its):
    CHIS = np.linspace(0, 1)
    num_spread = []
    
    for CHI in CHIS:
        int_corr_num = np.array(ods) + CHI*np.array(its)
        num_spread.append(np.std(int_corr_num))
    return CHIS[np.argmin(num_spread)]
```

```python
opt_chis = []
opt_nums = []
for image_time in image_times:
    this_chi = get_optimal_chi(od_dict[image_time], it_dict[image_time])
    opt_chis.append(this_chi)
    this_num = np.mean(np.array(od_dict[image_time]) + this_chi * np.array(it_dict[image_time]))
    opt_nums.append(this_num)
plt.plot(np.array(image_times)*1e6, opt_chis, 'o', markersize=20)
plt.xlim(0, 100)
plt.figure()
plt.plot(np.array(image_times)*1e6, opt_nums, 'o', markersize=20)
```

```python
# check if light counts are linear in image time
lc_data = []
for this_file in dist.filelist:
    this_img = CI(this_file)
    this_img.truncate_image(*window)
    this_light_counts = np.sum(np.int32(this_img.light_image_trunc) - np.int32(this_img.dark_image_trunc))
    this_time = this_img.get_image_time()
    lc_data.append((this_time, this_light_counts))
times, l_counts = zip(*lc_data)
```

```python
plt.plot(np.array(times)*1e6, l_counts, 'o')
```

```python
from collections import defaultdict
lc_dict = defaultdict(list)
for datum in lc_data:
    time, count = datum
    lc_dict[time].append(count)
    
mean_data = [(time, np.mean(lc_dict[time])) for time in sorted(lc_dict.keys())]
std_data = [(time, np.std(lc_dict[time])) for time in sorted(lc_dict.keys())]
times, mean_counts = zip(*mean_data)
times, noise_counts = zip(*std_data)
plt.errorbar(np.array(times)*1e6, mean_counts, yerr=noise_counts, marker='o', markersize=20)
plt.xlim(0,100)

plt.figure()
poissonian = np.array(noise_counts)**2 / np.array(mean_counts)
plt.plot(np.array(times)*1e6, poissonian, 'o', markersize=20)
```
