```python
mean_img =(0.5 * (np.int32(img.atom_image_trunc) + img.fluc_cor*np.int32(img.light_image_trunc))-np.int32(img.dark_image_trunc))
plt.imshow(mean_img)
plt.colorbar()
```

```python
diff = (img.fluc_cor*np.int32(img.light_image_trunc) - np.int32(img.atom_image_trunc))
print type(np.int32(img.light_image_trunc[0][0]))
```

```python
mean_img_hist, mean_img_bins, _ = plt.hist(np.ravel(mean_img),20)
plt.figure()
plt.hist(np.ravel(diff), 20)

print mean_img_bins,mean_img_hist

```

```python
diff_vec = np.ravel(diff)
mean_vec = np.ravel(mean_img)
diff_bin_labels = np.digitize(mean_vec, mean_img_bins)
bin_members = [[] for _ in range(len(mean_img_bins))]
for ii, this_diff in enumerate(diff_vec):
    try:
        bin_members[diff_bin_labels[ii]-1].append(this_diff)
    except IndexError:
        print diff_bin_labels[ii]
        raise
bin_vars = [np.std(members)**2 for members in bin_members]
```

```python
threshold = 0
filtered_bin_vars = [0 if meancounts < threshold else bin_var for meancounts, bin_var in zip(mean_img_hist, bin_vars)]
len(filtered_bin_vars)
```

```python
plt.plot(mean_img_bins[:-1], filtered_bin_vars, 'o',markersize=10,)
plt.plot(mean_img_bins, 2*mean_img_bins)
#plt.plot(mean_img_bins[:-1], mean_img_hist)
plt.xlabel('Mean Image Counts (Photons)')
plt.ylabel('Difference Image Variance (Photons)')
#plt.xlim(0,3000)
#plt.ylim(0,6000)
plt.title('Noise as a function of photon counts,VVA1.5')
plt.show()

```
