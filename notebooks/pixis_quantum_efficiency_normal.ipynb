```python
dist = CD(datadir, False)
true_power = 5.1e-6
photon_energy = 1.589 * 1.602e-19 #transition energy in eV, from Steck, times electron charge
qes = []

for fname in dist.filelist:
    img = CI(fname)
    img.truncate_image(450, 640, 800, 1000)
    counts = np.sum(np.int32(img.light_image_trunc) - np.int32(img.dark_image_trunc))
    image_time = img.get_image_time()
    
    true_energy = true_power * image_time
    true_counts = true_energy / photon_energy
    
    qe = counts / true_counts
    qes.append(qe)
    print qe
print '-'*40
print 'mean qe: \t%2.0f %%'%(np.mean(qes)*100)
print 'std: \t\t%2.1f'%(np.std(qes)*100)
```
