```python
from scipy.optimize import curve_fit
from scipy.ndimage import filters

def harm_func(x, A, B, C):
    return A*(x - B)**2 + C

def fit_harm(img):
    if np.mean(filters.gaussian_filter(img, 1, order=2)) < 0: #hump
        C0 = np.max(img)
        B0 = np.argmax(img)
        A0 = (img[0] - C0) / B0**2
    else:
        C0 = np.min(img)
        B0 = np.argmin(img)
        A0 = (img[0] - C0) / B0**2
    p0 = [A0, B0, C0]
    exes = np.array(range(len(img)))
    coef, _ = curve_fit(harm_func, exes, img, p0 = p0)
    return coef
```

```python
harm_coef_gpe = fit_harm(gpe_field)
exes = np.array(range(len(gpe_field)))
plt.plot(gpe_field*1e9)
plt.plot(harm_func(exes, *harm_coef_gpe)*1e9)

harm_coef_gerb = fit_harm(gerb_field)
plt.plot(gerb_field*1e9)
plt.plot(harm_func(exes, *harm_coef_gerb)*1e9)

f_gpe_harm = sqrt(2*MU_B*harm_coef_gpe[0] / MM) / (2*pi) / real_pix
f_gerb_harm = sqrt(2*MU_B*harm_coef_gerb[0] / MM) / (2*pi) / real_pix

print 'GPE: \t\t%2.0f Hz'%f_gpe_harm
print 'Gerbier: \t%2.0f Hz'%f_gerb_harm
```
