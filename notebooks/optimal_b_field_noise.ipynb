```python
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
plt.rcParams['figure.figsize'] = (20.0, 10.0)
plt.rcParams['axes.labelsize'] = 22
plt.rcParams['axes.titlesize'] = 26
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['lines.linewidth'] = 3
```

```python
SSS = 330 #counts

HBAR = 1.05e-34
omega_perp = 2*pi*800
MU_B = 9.27e-24
A_SCAT = 5e-9
px_size = (13.0e-6 / 24)
XSECTION = 1.5e-13

B0 = HBAR * omega_perp / (2*MU_B)
print B0*1e9
CC = px_size / (4*A_SCAT)
print CC
ALPHA = px_size**2 / XSECTION
print ALPHA
```

```python
def bee_field(num):
    return 2*B0 * (np.sqrt(1+num / CC)-1)

dees = np.linspace(0, 10)
nums = dees*ALPHA
plt.plot(dees, bee_field(nums)*1e9)
plt.xlabel('Optical Depth')
plt.ylabel('B Field (nT)')
```

```python
def dee_var(ell):
    # with no atoms
    return (np.sqrt(2.0 / ell) + np.sqrt(2.0 * ell) / SSS)**2

def num_var(ell):
    return dee_var(ell)*ALPHA**2

def bee_var(ell):
    return (B0/CC)**2 * num_var(ell)

sats = np.linspace(0.01, 10, 100)
ells = SSS*sats

b_noise = np.sqrt(bee_var(ells))

plt.plot(sats, b_noise*1e9)
plt.ylim(0, np.max(b_noise*1e9))
plt.xlabel('Saturation')
plt.ylabel('Field Noise (nT)')
plt.title('Field Noise, no atoms, 10 usec')

print np.min(b_noise*1e9)
```

## with atoms

```python
from scipy import optimize

def passed_light(ell, dee):
    test_func = lambda x: np.log(ell / x) + (ell - x) / SSS - dee    
    return optimize.brentq(test_func, 1e-12, ell+1e-12)

def passed_light_vec(ell_arr, dee_arr):
    result = np.zeros(ell_arr.shape)
    for index, ell in np.ndenumerate(ell_arr):
        result[index] = passed_light(ell, dee_arr[index])
    return result

def dee_var(ell, dee):
    light_out = passed_light_vec(ell, dee)
    return (1.0 / ell + 1.0 / SSS)**2 * ell + (1.0 / light_out + 1.0 / SSS)**2 * light_out

def num_var(ell, num):
    return dee_var(ell, num / ALPHA)*ALPHA**2

def bee_var(num, ell):
    return (B0/CC)**2 * (num_var(ell, num) + num) / (1 + num / CC)
```

```python
def passed_light_1dee(ells, dee):
    ayys = []
    for ell in ells:
        test_func = lambda x: np.log(ell / x) + (ell - x) / SSS - dee    
        this_ayy = optimize.brentq(test_func, 1e-12, ell+1e-12)
        ayys.append(this_ayy)
    return np.array(ayys)

def dee_var_1dee(ell, dee):
    light_out = passed_light_1dee(ell, dee)
    return (1.0 / ell + 1.0 / SSS)**2 * ell + (1.0 / light_out + 1.0 / SSS)**2 * light_out

def num_var_1dee(ell, num):
    return dee_var_1dee(ell, num / ALPHA)*ALPHA**2

def bee_var_1dee(num, ell):
    return (B0/CC)**2 * (num_var_1dee(ell, num) + num) / (1 + num / CC)

sats_1d = np.linspace(0.5, 4, 100)
ells = SSS*sats_1d
dee = 3.0
num = dee*ALPHA
b_noise = np.sqrt(bee_var_1dee(num, ells))

plt.plot(sats_1d, b_noise*1e9)
print np.min(b_noise)*1e9
```

```python
sats_1d = np.linspace(0.5, 3, 100)
dees_1d = np.linspace(0.01, 4, 100)

sats, dees = np.meshgrid(sats_1d, dees_1d)

ells = SSS*sats
nums = dees * ALPHA

b_noise = np.sqrt(bee_var(nums, ells))

plt.contourf(dees, sats, b_noise*1e9, 50)
plt.xlabel('Optical Depth')
plt.ylabel('Saturation')
plt.colorbar()

#plt.plot(sats_1d, np.ravel(b_noise)*1e9)

print np.min(b_noise*1e9)
```
