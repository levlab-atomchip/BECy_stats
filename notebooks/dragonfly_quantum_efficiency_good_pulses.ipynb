```python
def dragonfly_qe(img):
    image_time = img.get_image_time()
    photon_energy = 1.589 * 1.602e-19 #transition energy in eV, from Steck, times electron charge
    img.truncate_image(*WINDOW)
    diff_img = np.int32(img.light_image_trunc) - np.int32(img.dark_image_trunc)
    counts = np.sum(diff_img) / 16

    beam_energy = counts * photon_energy
    beam_power = beam_energy / image_time

    qe = beam_power / TRUE_POWER
    return qe
```
