```python
allsats=[]
#for dist in alldists:
sats_1 = []
for this_file in dist.filelist:
    this_img = CI(this_file)
    if this_img.curr_cont_par > 1.6:
        this_img.truncate_image(*window_1)
        sats_1.append(this_img.saturation())
allsats.append(sats_1)
```

```python
opt_dens_in = []
int_term_in = []
#for dist in alldists:
for this_file in dist.filelist:
    this_img = CI(this_file)
    if this_img.curr_cont_par > 1.6:
        this_img.truncate_image(*window_1)
        opt_dens_in.append(this_img.optdens_number(axis=0))
        int_term_in.append(this_img.int_term_number())
```

```python
plt.plot(np.ravel(allsats), opt_dens_in, 'o', markersize=15)
plt.plot(np.ravel(allsats), int_term_in, 'o', markersize=15)
plt.ylim(0, 40000)
```

```python
tot_num = np.array(opt_dens_in) + np.array(int_term_in)

#tot_num = filter(lambda x: x < 1e5, tot_num)
plt.plot(np.ravel(allsats), tot_num, 'o', markersize=15)
plt.ylim(0, 40000)
```

```python
#ALPHAS = [1, 1.5, 2, 2.5, 3]
#ALPHAS = [.01, .02, .05, .1, .2, .3, .4, .5, .6]
ALPHAS = np.linspace(0, 2)
num_spread = []
for ALPHA in ALPHAS:
    int_corr_num = ALPHA*np.array(opt_dens_in) + np.array(int_term_in)
    int_corr_num = filter(lambda x: x < 1e5, int_corr_num)
    #plt.plot(np.ravel(allsats), int_corr_num, 'o', markersize=15, label='%2.1f'%ALPHA)
    num_spread.append(np.std(int_corr_num))
#plt.legend(loc='upper right')

plt.figure()
plt.plot(ALPHAS, num_spread)
opt_alpha = ALPHAS[np.argmin(num_spread)]
print 'Optimal Alpha: %2.2f'%opt_alpha

plt.figure()
opt_int_corr_num = opt_alpha*np.array(opt_dens_in) + np.array(int_term_in)
plt.plot(np.ravel(allsats), opt_int_corr_num, 'o', markersize=15, label='%2.1f'%ALPHA)
plt.ylim(0, 4e4)
print 'Intensity-Corrected Number: %d'%int(round(np.mean(opt_int_corr_num)))
```
