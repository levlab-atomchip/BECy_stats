## Array of RectWires

```python
wa_num = 39
wa_period = 20e-6
wa_width = 2.0e-6
wa_height = 400e-9
wa_big_width = 200e-6
wa_grid_width = 800e-6
wa_length = 10e-6
wa_pos = np.cumsum(np.ones(wa_num)*wa_period) - 0.5 * wa_grid_width
```

```python
wa_current_total = 2e-3
wa_width_total = wa_num * wa_width + 2*wa_big_width
wa_current_sm = wa_width / wa_width_total * wa_current_total
wa_current_big = wa_big_width / wa_width_total * wa_current_total
print wa_current_sm
print wa_current_big
wa_width_full = 2*wa_big_width + wa_grid_width
```

```python
vwires = []
for nn, ypos in enumerate(wa_pos):
    vwires.append(acw.HRectWire('Grid Wire %d'%nn,
                                length = wa_length,
                                width = wa_width,
                                height = wa_height,
                                current = wa_current_sm,
                                x0 = -0.5*wa_length,
                                y0 = ypos,
                                z0 = 0))
vwires.append(acw.HRectWire('Big Wire 1',
                                length = wa_length,
                                width = wa_big_width,
                                height = wa_height,
                                current = wa_current_big,
                                x0 = -0.5*wa_length,
                                y0 = -0.5 * wa_grid_width - wa_big_width,
                                z0 = 0))
vwires.append(acw.HRectWire('Big Wire 2',
                                length = wa_length,
                                width = wa_big_width,
                                height = wa_height,
                                current = wa_current_big,
                                x0 = -0.5*wa_length,
                                y0 = 0.5 * wa_grid_width,
                                z0 = 0))
vwires.append(acw.HRectWire('Full Wire Neg',
                                length = 300e-6,
                                width = wa_width_full,
                                height = wa_height,
                                current = wa_current_total,
                                x0 = -0.5 * wa_length - 300e-6,
                                y0 = -0.5 * wa_width_full,
                                z0 = 0))
vwires.append(acw.HRectWire('Full Wire Pos',
                                length = 300e-6,
                                width = wa_width_full,
                                height = wa_height,
                                current = wa_current_total,
                                x0 = 0.5 * wa_length,
                                y0 = -0.5 * wa_width_full,
                                z0 = 0))
```

```python
atom_height = cloud_height
```

```python
wa_locs_y = np.linspace(-1e-3, 1e-3, 3000)
wa_fields_y = np.array([-1*np.sum([ww.bfieldcalc(0, ll, atom_height)[1] for ww in vwires]) for ll in wa_locs_y])
plt.plot(wa_locs_y*1e6, wa_fields_y*1e9)
plt.xlabel('Height (um)')
plt.ylabel('Field (nT)')
plt.title('Field barrier across wire')
```

```python
def N_1d_from_mu(mu_guess, tot_field, pxsize = 400e-6 / 1024):
    '''expects mu as field in gauss'''
    mu_guess *= (acm.MU_B * 1e-4)
    n_1d = np.array([(0.25 / acm.A) * ((np.min([acm.MU_B * ff - mu_guess, 0]) / (acm.HBAR * tight_freq))**2 - 1) for ff in tot_field])
    N_1d = [nn*pxsize if nn > 0 else 0 for nn in n_1d]
    return N_1d

def N_1d_from_N(target_N, tot_field, pxsize = 400e-6 / 1024):
    min_func_N = lambda x: (N_from_mu(x, tot_field) - target_N)**2
    optres = optimize.minimize_scalar(min_func_N, method = 'Golden', bracket = (np.min(tot_field)*1e4, np.max(tot_field)*1e4))
    return N_1d_from_mu(optres.x, tot_field)
```
