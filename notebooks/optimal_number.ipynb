```python
import matplotlib.cm as cm

ells_1d = np.linspace(0.001, 3, 100) # incident light
ayys_1d  = np.linspace(0.001, 0.4, 100) # passed light
sss = 1 # saturation

ells, ayys = np.meshgrid(ells_1d, ayys_1d)

numer = (np.log(ells / ayys) + (ells - ayys)/sss)**2
denom = (1 / ells + 1 / sss)**2 * ells + (1 / ayys + 1 / sss)**2 * ayys
snr = np.sqrt(numer / denom)
snr_counts = 330 # at 10 usec

plt.contourf(ells, ayys, snr*snr_counts, 50, cmap=cm.Blues)
plt.colorbar()

trial_ds = np.linspace(2.2, 3.2, 5)
for ii, dee in enumerate(trial_ds):
    this_opt_A_low = ells_1d*np.exp(-dee)
    this_opt_A_high = ells_1d - dee
    this_opt_A_approx = np.where(this_opt_A_low > this_opt_A_high, this_opt_A_low, this_opt_A_high)
    plt.plot(ells_1d, this_opt_A_approx, linewidth=2, label='D=%2.1f'%dee)
plt.ylim(0, 0.4)

plt.legend(loc='upper left')
```

```python
def passed_light(applied_light, optical_depth):
    result = []
    for this_light in applied_light:
        test_func = lambda x: np.log(this_light / x) + this_light - x - optical_depth    
        result.append(optimize.brentq(test_func, 1e-6, this_light-1e-6))
    return np.array(result)
```

```python
test_ells = np.linspace(0.001, 10.0, 50)
test_dee = 2.2
test_ayys = passed_light(test_ells, test_dee)
plt.plot(test_ells, test_ayys/test_ells, label='Full Solution')
plt.plot(test_ells, np.ones(len(test_ells))*np.exp(-test_dee), label='Low Intensity')
plt.plot(test_ells, (test_ells - test_dee) / test_ells, label='High Intensity')

plt.ylim(0, 1)
plt.xlabel('Saturation')
plt.ylabel('Passed Light (%)')
plt.legend(loc = 'upper left')
```

```python
max_ell = 10
ells_1d = np.linspace(0.001, max_ell, 100) # incident light
ayys_1d  = np.linspace(0.001, max_ell, 100) # passed light
sss = 1 # saturation

ells, ayys = np.meshgrid(ells_1d, ayys_1d)

numer = (np.log(ells / ayys) + (ells - ayys)/sss)**2
denom = (1 / ells + 1 / sss)**2 * ells + (1 / ayys + 1 / sss)**2 * ayys
snr = np.sqrt(numer / denom)
snr_counts = 330 # at 10 usec

plt.contourf(ells, ayys, snr*snr_counts, 50, cmap=cm.Blues, vmax=600)
plt.colorbar()

trial_ds = [1, 2, 3, 4, 5]
for ii, dee in enumerate(trial_ds):
    try:
        this_A = passed_light(ells_1d, dee)
    except ValueError:
        this_light = 1
        test_func = lambda x: np.log(this_light / x) + this_light - x - dee
        plt.figure()
        plt.plot(ells_1d, test_func(ells_1d))
        raise ValueError
    plt.plot(ells_1d, this_A, linewidth=2, label='D=%2.1f'%dee)
    plt.ylim(0, max(ayys_1d))
plt.plot(ells_1d, ells_1d, linewidth=5, color='black')

plt.legend(loc='upper left')
```

```python
def snr(sat, optical_depth):
    sss = 1
    light_out = passed_light(sat, optical_depth)
    numer = (np.log(sat / light_out) + (sat - light_out)/sss)**2
    denom = (1 / sat + 1 / sss)**2 * sat + (1 / light_out + 1 / sss)**2 * light_out
    return np.sqrt(numer / denom)
```

```python
sats = np.linspace(1e-3, 4, 50)
dees = [0.5, 1, 2, 3]
for dee in dees:
    plt.plot(sats, snr(sats, dee)*snr_counts, label='OD = %2.1f'%dee)
plt.xlabel('Saturation')
plt.ylabel('SNR')
plt.legend(loc='upper left')
```

```python
def optimal_sat(optical_depth):
    test_func = lambda sat: -snr(np.array([sat]), optical_depth)
    result = optimize.minimize_scalar(test_func, bounds=(0, 10), method='bounded')
    return result.x[0][0][0][0][0][0] #I don't know why this is 6-fold nested.
```

```python
dees = np.linspace(0.01, 3, 50)
opt_sats = map(optimal_sat, dees)
plt.plot(dees, opt_sats, linewidth=10)
plt.xlabel('Optical Depth')
plt.ylabel('Optimal Saturation')
plt.ylim(0, max(opt_sats))
```

```python
opt_snrs = np.array([snr(np.array([sat]), dee) for sat, dee in zip(opt_sats, dees)])
plt.plot(dees, opt_snrs*snr_counts, linewidth=10)
plt.xlabel('Optical Depth')
plt.ylabel('Optimal SNR')
```
