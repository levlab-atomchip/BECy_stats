```python
alldists=[dist_15,dist_17,dist_19, \
              dist_22,dist_235,dist_25, \
              dist_26]
allfiles=np.vstack(tuple([dist_15.filelist,dist_17.filelist,dist_19.filelist, \
                          dist_22.filelist,dist_235.filelist,dist_25.filelist, \
                          dist_26.filelist]))
len(alldists)
```

```python
img_1 = CI(allfiles[6,0])
plt.imshow(img_1.get_od_image(trunc_switch=False))
```

```python
img_1.truncate_image(580, 900, 500, 800)
plt.figure()
plt.imshow(img_1.get_od_image())
window_1 = [580, 900, 500, 800]
```

```python
for dist in alldists:
    dist.custom_fit_window = window_1
```

```python
#icnums_in = dist_in.get_ic_atom_numbers()

all_icnums=[]
for dist in alldists:
    icnums_1 = dist.get_ic_atom_numbers()
    all_icnums.append(icnums_1)
```

```python
allsats=[]
for dist in alldists:
    sats_1 = []
    for this_file in dist.filelist:
        this_img = CI(this_file)
        this_img.truncate_image(*window_1)
        sats_1.append(this_img.saturation())
    allsats.append(sats_1)
```

```python
#plt.plot(sats_in, icnums_in, '*', markersize=15, label='In Situ')
for sats, icnums in zip(allsats,all_icnums):
    plt.plot(sats, icnums, 'o', markersize=15, label='TOF 5 ms')
plt.xlabel('Saturation')
plt.ylabel('Number')
```

```python
opt_dens_in = []
int_term_in = []
for dist in alldists:
    for this_file in dist.filelist:
        this_img = CI(this_file)
        this_img.truncate_image(*window_1)
        opt_dens_in.append(this_img.optdens_number())
        int_term_in.append(this_img.int_term_number())
```

```python
plt.plot(np.ravel(allsats), opt_dens_in, 'o', markersize=15)
plt.plot(np.ravel(allsats), int_term_in, 'o', markersize=15)
```

```python
#ALPHAS = [1, 1.5, 2, 2.5, 3]
ALPHAS = [.01, .02, .05, .1, .2, .3, .4, .5, .6]
ALPHAS = np.linspace(0, 1)
num_spread = []
for ALPHA in ALPHAS:
    int_corr_num = ALPHA*np.array(opt_dens_in) + np.array(int_term_in)
    #plt.plot(np.ravel(allsats), int_corr_num, 'o', markersize=15, label='%2.1f'%ALPHA)
    num_spread.append(np.std(int_corr_num))
#plt.legend(loc='upper right')

plt.figure()
plt.plot(ALPHAS, num_spread)
opt_alpha = ALPHAS[np.argmin(num_spread)]
print 'Optimal Alpha: %2.2f'%opt_alpha

plt.figure()
opt_int_corr_num = opt_alpha*np.array(opt_dens_in) + np.array(int_term_in)
plt.plot(np.ravel(allsats), opt_int_corr_num, 'o', markersize=15, label='%2.1f'%ALPHA)
print 'Intensity-Corrected Number: %d'%int(round(np.mean(opt_int_corr_num)))
```

```python
#CHIS = [.01, .02, .05, .1, .2, .3, .4, .5, .6]
CHIS = np.linspace(0, 10)
num_spread = []
for CHI in CHIS:
    int_corr_num = np.array(opt_dens_in) + CHI*np.array(int_term_in)
    #plt.plot(np.ravel(allsats), int_corr_num, 'o', markersize=15, label='%2.1f'%ALPHA)
    num_spread.append(np.std(int_corr_num))
#plt.legend(loc='upper right')

plt.figure()
plt.plot(CHIS, num_spread)
opt_chi = CHIS[np.argmin(num_spread)]
print 'Optimal Chi: %2.2f'%opt_chi

plt.figure()
opt_int_corr_num = np.array(opt_dens_in) + opt_chi* np.array(int_term_in)
plt.plot(np.ravel(allsats), opt_int_corr_num, 'o', markersize=15)
print 'Intensity-Corrected Number: %d'%int(round(np.mean(opt_int_corr_num)))
```
