```python
bfs = [] #list of reconstructed magnetic fields from single rows of pixels
for img in imgs:
    img.truncate_image(396,623,140,143)
    bfs.append(img.get_gerbier_field(filter_on = True))
for bf in bfs:
    plt.plot(bf*1e9)
```

```python
allruns = np.vstack(tuple(bfs))
noise=np.std(allruns,axis = 0)
plt.plot(noise*1e9)
print np.mean(noise*1e9)
print np.std(noise*1e9)/np.sqrt(len(noise))
```
