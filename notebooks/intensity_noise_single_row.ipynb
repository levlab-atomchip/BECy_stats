```python
dist = CDna(datadir, False)
imgs = [] #list of all images in the distribution
for file in dist.filelist:
    imgs.append(nai(file))
```

```python
#define region of interest
xmin=396
xmax=623
ymin=100
ymax=250
```

```python
bfs = [] #list of reconstructed magnetic fields from single rows of pixels
for img in imgs:
    img.truncate_image(xmin,xmax,140,140+1)
    bfs.append(img.get_gerbier_field(filter_on = True))
for bf in bfs:
    plt.plot(bf*1e9)
```

```python
def calc_field_noise(bfs):
    #calculate field noise from a list of reconstructed fields, returns the average noise and its error
    allruns = np.vstack(tuple(bfs))
    noise=np.std(allruns,axis = 0)
    #plt.plot(noise*1e9)
    #plt.xlabel('pixel number')
    #plt.ylabel('field noise (nT)')
    return np.mean(noise*1e9), np.std(noise*1e9)/np.sqrt(len(noise))
```

```python
calc_field_noise(bfs)
```

Now I want to use different regions of interest to calculate field noise and see how thst changes as a function of the number of rows used in the region of interest.

```python
def gen_y_lims(n_row):
    y1 = np.random.randint(ymin,ymax+1)
    if y1+n_row>ymax:
        y2=ymax-n_row
        return [y2,ymax]
    else:
        y2=y1+n_row
        return [y1,y2]
```

```python
def get_b_field(imgs,n_row,y_lims=None):
    if n_row > ymax-ymin:
        print 'too many rows for the ROI'
        print 'try something smaller than or equal to ' + str(ymax-ymin)
        return
    bfs = []
    if y_lims == None:
        ylims=gen_y_lims(n_row)
    else:
        ylims=y_lims
    for img in imgs:
        img.truncate_image(xmin,xmax,*ylims)
        bfs.append(img.get_gerbier_field(filter_on = True))
    #for bf in bfs:
        #plt.plot(bf*1e9)
    return bfs,ylims
```

```python
bfs,_=get_b_field(imgs,10)
```

```python
noises = []
errs_noise = []
n_rows = range(ymax-ymin)+np.ones(ymax-ymin)
for nn in n_rows:
    bfs,_ = get_b_field(imgs,nn)
    ns,err = calc_field_noise(bfs)
    noises.append(ns)
    errs_noise.append(err)
```

```python
plt.errorbar(n_rows,noises,yerr=errs_noise,fmt='o')
plt.plot(n_rows,noises[-1]/np.sqrt((ymax-ymin)/n_rows))
plt.xlabel('number of rows in ROI')
plt.ylabel('reconstructed field noise')
plt.title('VVA1.8, 10usec expose time, no atoms (01/06/2015)')
```

```python
noises_y = []
errs=[]
y_range = range(ymax-ymin)
ys=[]
for ii in y_range:
    bfs,ylims = get_b_field(imgs,1,[ymin+ii,ymin+ii+1])
    ys.append(ylims[0])
    ns,err = calc_field_noise(bfs)
    noises_y.append(ns)
    errs.append(err)
plt.errorbar(ys,noises_y,yerr=errs,fmt='o')
print np.mean(errs)
print np.std(noises_y)
```
