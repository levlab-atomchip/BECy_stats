```python
from scipy.integrate import trapz
```

```python
# http://www.variousconsequences.com/2010/01/fft-based-abel-inversion-tutorial.html
def abel(dfdx, x): 
    nx = len(x) 
    # do each half of the signal separately (they should be the same 
    # up to noise) 
    integral = np.zeros((2,nx/2), dtype=float) 
    for i in xrange(nx/2, nx-1): 
        divisor = np.sqrt(x[i:nx]**2 - x[i]**2) 
        integrand = dfdx[i:nx] / divisor 
        integrand[0] = integrand[1] # deal with the singularity at x=r 
        integral[0][i-nx/2] = - trapz(integrand, x[i:nx]) / pi 
    for i in xrange(nx/2, 1, -1): 
        divisor = np.sqrt(x[i:0:-1]**2 - x[i]**2) 
        integrand = dfdx[i:0:-1] / divisor 
        integrand[0] = integrand[1] # deal with the singularity at x=r 
        integral[1][-i+nx/2] = - trapz(integrand, x[i:0:-1]) / pi 
    return(integral)
```

```python
def centroid(arr):
    #cdf = np.cumsum(arr)
    #total = np.sum(arr)
    #middle = 0.5*total
    #return np.argmin((cdf - middle)**2)
    return np.argmax(arr)

def recenter(arr):
    this_centroid = centroid(arr)
    halfwidth = min([this_centroid, len(arr) - this_centroid - 1])
    return arr[this_centroid - halfwidth: this_centroid + halfwidth + 1]
```

## Slice by Slice Abel

```python
def column_abel(column, arrs=y1d):
    dee_arr = arrs[1]-arrs[0]
    pos_arrs = np.array(filter(lambda x: x>=0, arrs))[:-1]
    column_prime = filters.gaussian_filter(column, 1, order=1) / dee_arr
    column_abel = abel(column_prime, arrs)[1,:]
    
    #column_abel_norm = 2*pi*np.sum(column_abel*pos_arrs*dee_arr)
    #column_abel /= column_abel_norm
    return column_abel

def slicing_abel(cd_img):
    column_abels = []
    for col_idx in range(cd_img.shape[1]):
        column_abels.append(column_abel(cd_img[:, col_idx]))
    return np.transpose(np.vstack(tuple(column_abels)))
```
