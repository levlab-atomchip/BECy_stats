```python
def sq_parabola(x, AA, mu, ww):
    return (AA**0.5-(mu-x)**2/ww**2)**2
```

```python
def fit_sq_parabola(img):
    exes = np.array(range(len(img)))
    AA = np.max(img)
    mu = np.argmax(img)
    ww = len(img) / 2
    coefs, _ = curve_fit(sq_parabola, exes, img, p0=[AA, mu, ww])
    return coefs
```
