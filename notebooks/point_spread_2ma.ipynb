```python
test_od_roi = test_od[182:195, 217:931]
```

```python
xaxis = np.cumsum(np.ones(test_od_roi.shape[1])*PIXSIZE)
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
plt.plot(xaxis, np.sum(test_od_roi, axis=0))
plt.xlabel('Position / um')
plt.ylabel('Density / a.u.')
```

```python
imgs_2 = [ci(ff) for ff in dist_2.filelist]
cdimgs_2 = [im.get_od_image(True, False)[182:195, 217:931] / im.s_lambda for im in imgs_2]
ldimgs_2 = [bp.line_density(cdim, PIXSIZE) for cdim in cdimgs_2]
```

```python
ldsnorm_2 = [np.roll((ld-np.mean(ld))/np.std(ld), 0) for ii, ld in enumerate(ldimgs_2)]
fig = plt.figure()
fig.set_size_inches(20,10)
for ld in ldsnorm_2[:5]:
    plt.plot(xaxis,ld)
```

```python
test_ld = ldimgs_2[0]
```

```python
test_norm = (test_ld - np.mean(test_ld)) / np.std(test_ld)
test_fft = np.fft.fftshift(np.fft.fft(test_norm, 512))
fig = plt.figure()
fig.set_size_inches(20,10)
plt.semilogy((np.abs(test_fft[256:])**2))
```

```python
fft_2 = [np.fft.fftshift(np.fft.fft(ld, 512)) for ld in ldsnorm_2]
psd_2 = [np.abs(ft)**2 for ft in fft_2]
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(512, PIXSIZE))
for dd in psd_2:
    plt.semilogy(1.0 / faxis[258:350],dd[258:350])
    plt.xlim((0,40))
```

```python
psd_avg_2 = np.mean(psd_2, axis=0)
psd_norm_2 = psd_avg_2 / sum(psd_avg_2)
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(512, PIXSIZE))

plt.semilogy(1.0 / faxis[256:], psd_norm_2[256:])
plt.xlim((0,40))
```
