```python
shifts = np.array(range(-30, 30))
msds = np.zeros(len(shifts))
for ii, sh in enumerate(shifts):
    s1_shift = np.roll(s1, sh)
    if sh != 0:
        msds[ii] = np.mean(((s1_shift - s2)**2)[abs(sh):-abs(sh)])
    else:
        msds[ii] = np.mean((s1_shift - s2)**2)
plt.plot(shifts * 13.0 / 21.1, msds, '.')
plt.xlabel('Shift / um')
plt.ylabel('Mean Square Difference')
bestshift = shifts[np.argmin(msds)]*13.0 / 21.1
print 'Optimal shift is %2.1f um'%bestshift
```

```python
PIXSIZE = 13.0 / 21.1

def optimal_shift(img, ref_img):
    s1 = img
    s2 = ref_img
    shifts = np.array(range(-30, 30))
    msds = np.zeros(len(shifts))
    for ii, sh in enumerate(shifts):
        s1_shift = np.roll(s1, sh)
        if sh != 0:
            msds[ii] = np.mean(((s1_shift - s2)**2)[abs(sh):-abs(sh)])
        else:
            msds[ii] = np.mean((s1_shift - s2)**2)
    bestshift = shifts[np.argmin(msds)]*PIXSIZE
    #plt.plot(shifts, msds)
    return bestshift
```

```python
GET_SHIFT_RE = r'/home/will/levlab/data/shiftstage/good_set/([\d.]*)/.*\.mat'

import re

def get_shift(filename):
    p = re.compile(get_shift_re)
    return p.match(filename).group(1)

if p.match(d1_5.filelist[0]) is not None:
    print get_shift(d1_5.filelist[0])
else:
    print 'No Match'
        
```

```python
dref = CD(dir + '1.505/', False)
avgref = dref.get_average_image()
sref = np.sum(avgref, axis=0) # Reference data
plt.plot(sref)
```

```python
import glob
filelist = glob.glob(dir + '*/*.mat')
shifts = np.zeros(len(filelist))
meas_shifts = np.zeros(len(filelist))
for ii, filename in enumerate(filelist):
    #print filename
    shift = (float(get_shift(filename)) - 1.505) * 1000
    img = ci(filename).get_od_image()
    side = np.sum(img, axis=0)
    side /= np.sum(side)
    side *= len(side)
    shifts[ii] = shift
    meas_shifts[ii] = optimal_shift(side, sref)
plt.plot(shifts, meas_shifts, '.')
slope, intercept, _, _, _ = scipy.stats.linregress(shifts, meas_shifts)
print 'Slope is %2.2f'%slope
predicted = slope * shifts + intercept
resid = predicted - meas_shifts
print 'Resolution is %2.1f um'%np.std(resid)
```
