## Ideal Response

```python
pxsize = 13.0 / 24
print 2.0 / pxsize
```

```python
per_px = 37
comb_20 = np.array([0 if (nn%per_px > 3) else 1 for nn in range(714)])
```

```python
comb_fft = np.fft.fftshift(np.fft.fft(comb_20, 1024))
comb_psd = np.abs(comb_fft[513:])**2
comb_psd /= np.sum(comb_psd) / len(comb_psd) #normalize
```

```python
import scipy.signal
```

```python
gauss_env = scipy.signal.gaussian(714, 50 / PIXSIZE)
```

```python
bfield = comb_20 * gauss_env
xaxis = np.cumsum(np.ones(714)*pxsize)
plt.plot(xaxis, bfield)
plt.xlabel('Position / um')
plt.ylabel('B Field / a.u.')
plt.title('Idealized Potential')
```

```python
bf_fft = np.fft.fftshift(np.fft.fft(bfield, 1024))
faxis = np.fft.fftshift(np.fft.fftfreq(1024, pxsize))
bf_psd = np.abs(bf_fft)**2
bf_psd /= np.sum(bf_psd) #normalize
plt.plot(faxis[513:], bf_psd[513:])
plt.xlabel('Wavenumber / um^-1')
plt.title('Power Spectrum of Idealized Potential')
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)

plt.semilogy(0.5 / faxis[513:], bf_psd[513:])
plt.semilogy(0.5 / faxis[513:], psd_norm[513:])

plt.xlim((0.5,5))
plt.ylim((1e-11, 1e-1))
plt.xlabel('Resolution / um')
plt.title('Resolution, idealized and data')
```
