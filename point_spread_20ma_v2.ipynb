```python
fft_psf = [np.fft.fftshift(np.fft.fft(ld, 1024)) for ld in ldsnorm]
psd = [np.abs(ft)**2 for ft in fft_psf]
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))
for dd in psd:
    plt.semilogy(1.0 / faxis[513:],dd[513:])
    plt.xlim((0,40))
```

```python
psd_avg = np.mean(psd, axis=0)
psd_norm = psd_avg / sum(psd_avg)
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))

plt.semilogy(1.0 / faxis[513:], psd_norm[513:])
plt.xlim((0,40))
```

## Ideal Response

```python
per_px = 37
comb_20 = np.array([0 if (nn%per_px != 0 and nn%per_px != 1) else 1 for nn in range(714)])
len(comb_20)
```

```python
plt.plot(comb_20)
```

```python
comb_fft = np.fft.fftshift(np.fft.fft(comb_20, 1024))
comb_psd = np.abs(comb_fft[513:])**2
comb_psd /= np.sum(comb_psd) / len(comb_psd) #normalize
plt.plot(comb_psd)
```

```python
import scipy.signal
```

```python
gauss_env = scipy.signal.gaussian(714, 50 / PIXSIZE)
plt.plot(gauss_env)
```

```python
bfield = comb_20 * gauss_env
plt.plot(bfield)
```

```python
bf_fft = np.fft.fftshift(np.fft.fft(bfield, 1024))
bf_psd = np.abs(bf_fft)**2
bf_psd /= np.sum(bf_psd) #normalize
plt.plot(bf_psd[513:])
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))

plt.semilogy(1.0 / faxis[513:], bf_psd[513:])
plt.semilogy(1.0 / faxis[513:], psd_norm[513:])

plt.xlim((0,40))
```
