```python
def N_from_mu(mu_guess, tot_field):
    ''' expects mu as a field in gauss'''
    mu_guess *= (acm.MU_B * 1e-4)
    n_1d = np.array([(0.25 / acm.A) * ((np.min([acm.MU_B * ff - mu_guess, 0]) / (acm.HBAR * tight_freq))**2 - 1) for ff in tot_field])
    N_1d = [nn*(400e-6/1024) if nn > 0 else 0 for nn in n_1d]
    return np.sum(N_1d)
```

```python
def N_1d_from_mu(mu_guess, tot_field, pxsize = 400e-6 / 1024):
    '''expects mu as field in gauss'''
    mu_guess *= (acm.MU_B * 1e-4)
    n_1d = np.array([(0.25 / acm.A) * ((np.min([acm.MU_B * ff - mu_guess, 0]) / (acm.HBAR * tight_freq))**2 - 1) for ff in tot_field])
    N_1d = [nn*pxsize if nn > 0 else 0 for nn in n_1d]
    return N_1d
```

```python
def N_1d_from_N(target_N, tot_field, pxsize = 400e-6 / 1024):
    min_func_N = lambda x: (N_from_mu(x, tot_field) - target_N)**2
    optres = optimize.minimize_scalar(min_func_N, method = 'Golden', bracket = (np.min(tot_field)*1e4, np.max(tot_field)*1e4))
    return N_1d_from_mu(optres.x, tot_field)
```

## Thermal Cloud

```python
def thermal_density(NN, TT, tot_field):
    expV = np.exp(-acm.MU_B * tot_field / (acm.K_B * TT))
    expV *= (NN / np.sum(expV))
    return expV
```

## Partial Condensates

```python
def critical_temp(NN, long_freq=2*acm.pi*10, tight_freq=2*acm.pi*1000):
    ''' Critical temperature for condensation of a gas of N atoms in a 3D harmonic oscillator '''
    wbar = (long_freq * tight_freq**2)**(1.0/3)
    return (0.94 * acm.HBAR * wbar * NN**(1.0/3)) / acm.K_B
```

```python
def condensate_fraction(TT, NN, long_freq = 2*acm.pi*10, tight_freq=2*acm.pi*1000):
    TC = critical_temp(NN, long_freq, tight_freq)
    return np.max([1 - (TT/TC)**3, 0])
```

```python
def partial_density(NN, TT, tot_field, long_freq = 2*acm.pi*10, tight_freq = 2*acm.pi*1000):
    cf = condensate_fraction(TT, NN, long_freq, tight_freq)
    N_BEC = cf*NN
    N_TH = (1-cf)*NN
    N_1D_TH = thermal_density(N_TH, TT, tot_field)
    N_1D_BEC = np.array(N_1d_from_N(N_BEC, tot_field))
    return N_1D_TH + N_1D_BEC
```
