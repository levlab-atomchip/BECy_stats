## Potential Reconstruction

```python
test_img = ci(dist.filelist[0])
test_cd = test_img.get_od_image(True, False)[174:193, 217:931] / test_img.s_lambda
test_ld = bp.line_density(test_cd, PIXSIZE)
xaxis = np.cumsum(np.ones(len(test_ld))*PIXSIZE)
```

```python
plt.plot(xaxis*1e6, test_ld*PIXSIZE)
plt.xlabel('Position / um')
plt.ylabel('Atoms / #')
plt.title('Example Line Density Data')
```

```python
coefs = fit_gaussian_1d(test_ld*PIXSIZE)
print coefs[2]*PIXSIZE*1e6
sigma_guess = 100e-6
plt.plot(xaxis*1e6, test_ld*PIXSIZE)
plt.xlabel('Position / um')
plt.ylabel('Atoms / #')
plt.title('Example Line Density Data')
plt.plot(xaxis*1e6, gaussian_1d(xaxis, 2.5, 145e-6, sigma_guess, 0, 0))
```

```python
temp_guess = 200e-9
omega_fit = np.sqrt(acm.K_B * temp_guess / (acm.M * sigma_guess**2))
f_fit = omega_fit / (2*acm.pi)
print f_fit
```

```python
temp = 200e-9
long_freq = 2*acm.pi*.1
center = 145e-6
V_recon = acm.K_B * temp * np.log(test_ld)
V_sample =0.5*acm.M*long_freq**2*(xaxis - center)**2 - V_recon
```

```python
plt.plot((xaxis-center)*1e6, ((V_sample) / acm.MU_B)*1e9)
plt.xlabel('Position / um')
plt.ylabel('Reconstructed Field / nT')
plt.title('Reconstruction of Magnetic Field from Thermal Cloud Data')
#plt.ylim(0, 75)
#print (np.max(V_sample) - np.min(V_sample)) / np.min(V_sample)
```
