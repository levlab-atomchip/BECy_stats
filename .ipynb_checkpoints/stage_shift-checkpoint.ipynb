```python
shifts = np.array(range(-30, 30))
msds = np.zeros(len(shifts))
for ii, sh in enumerate(shifts):
    s1_shift = np.roll(s1, sh)
    if sh != 0:
        msds[ii] = np.mean(((s1_shift - s2)**2)[abs(sh):-abs(sh)])
    else:
        msds[ii] = np.mean((s1_shift - s2)**2)
plt.plot(shifts * 13.0 / 21.1, msds, '.')
plt.xlabel('Shift / um')
plt.ylabel('Mean Square Difference')
bestshift = shifts[np.argmin(msds)]*13.0 / 21.1
print 'Optimal shift is %2.1f um'%bestshift
```

```python
PIXSIZE = 13.0 / 21.1

def optimal_shift(img, ref_img, shift_range=range(-30,30)):
    s1 = img
    s2 = ref_img
    shifts = np.array(shift_range)
    msds = np.zeros(len(shifts))
    for ii, sh in enumerate(shifts):
        s1_shift = np.roll(s1, sh)
        if sh != 0:
            msds[ii] = np.mean(((s1_shift - s2)**2)[abs(sh):-abs(sh)])
        else:
            msds[ii] = np.mean((s1_shift - s2)**2)
    bestshift = shifts[np.argmin(msds)]*PIXSIZE
    #plt.plot(shifts, msds)
    return bestshift
```

```python
GET_SHIFT_RE = r'/home/will/levlab/data/shiftstage/20141202_good_set/([\d.]*)/.*\.mat'

import re

def get_shift(filename, offset=1.505, multiplier=1000):
    p = re.compile(GET_SHIFT_RE)
    shift_str = p.match(filename).group(1)
    return (float(shift_str) - offset) * multiplier
```

```python
import glob
filelist = glob.glob(dir + '*/*.mat')
shifts = np.zeros(len(filelist))
meas_shifts = np.zeros(len(filelist))
for ii, filename in enumerate(filelist):
    #print filename
    shift = get_shift(filename)
    img = ci(filename).get_od_image()
    side = np.sum(img, axis=0)
    side /= np.sum(side)
    side *= len(side)
    shifts[ii] = shift
    meas_shifts[ii] = optimal_shift(side, sref)
plt.plot(shifts, meas_shifts, '.')
slope, intercept, _, _, _ = scipy.stats.linregress(shifts, meas_shifts)
print 'Slope is %2.2f'%slope
predicted = slope * shifts + intercept
resid = predicted - meas_shifts
print 'Resolution is %2.1f um'%np.std(resid)
```

```python
GET_SHIFT_RE2 = r'/home/will/levlab/data/shiftstage/20141204/([\w.]*)nm/.*\.mat'

import re

def get_shift2(filename, offset=800, multiplier=0.001):
    p = re.compile(GET_SHIFT_RE2)
    shift_str = p.match(filename).group(1)
    return (float(shift_str) - offset) * multiplier

p = re.compile(GET_SHIFT_RE2)
if p.match(d0nm.filelist[0]) is not None:
    print get_shift2(d0nm.filelist[0])
else:
    print 'No Match'
```

```python
import glob
import BECphysics as bp
import bfieldsensitivity as bf

PIXSIZE = 13.0/21.1

dref = cd.CD(dir2 + '0nm/', False)
avgref = dref.get_average_image(trunc_switch=False)[169:178, 350:600]
sample_img = ci(dref.filelist[0])
cdref = avgref / sample_img.s_lambda # Reference data
ldref = bp.line_density(cdref, PIXSIZE)
faref = bp.field_array(ldref)
#faref = bf.ci_to_fa(sample_img)
```

```python
def shift_resolution(dir2, fa_range=None):
    filelist = glob.glob(dir2 + '*/*.mat')
    shifts = np.zeros(len(filelist))
    meas_shifts = np.zeros(len(filelist))
    for ii, filename in enumerate(filelist):
        #print filename
        shift = get_shift2(filename, offset=0)
        if fa_range is not None:
            this_fa = bf.ci_to_fa(ci(filename))[fa_range]
        else:
            this_fa = bf.ci_to_fa(ci(filename))
        shifts[ii] = shift
        if shift >= 0:
            meas_shifts[ii] = optimal_shift(this_fa, faref, range(6))
        else:
            meas_shifts[ii] = optimal_shift(this_fa, faref, range(-5,1))
    plt.plot(shifts, meas_shifts, '.')
    plt.plot(shifts, shifts)
    slope, intercept, _, _, _ = scipy.stats.linregress(shifts, meas_shifts)
    print 'Slope is %2.2f'%slope
    predicted = slope * shifts + intercept
    resid = predicted - meas_shifts
    print 'Resolution is %2.1f um'%np.std(resid)
    plt.plot(shifts, predicted)
    #print shifts
    
shift_resolution(dir2, range(350, 600))
```
