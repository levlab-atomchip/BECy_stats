```python
imgs_5 = [ci(ff) for ff in dist_5.filelist]
cdimgs_5 = [im.get_od_image(True, False)[175:190, 217:931] / im.s_lambda for im in imgs_5]
ldimgs_5 = [bp.line_density(cdim, PIXSIZE) for cdim in cdimgs_5]
```

```python
ldsnorm_5 = [np.roll((ld-np.mean(ld))/np.std(ld), 0) for ii, ld in enumerate(ldimgs_5)]
fig = plt.figure()
fig.set_size_inches(20,10)
for ld in ldsnorm_5[:5]:
    plt.plot(xaxis,ld)
```

```python
test_ld = ldimgs_5[0]
```

```python
test_norm = (test_ld - np.mean(test_ld)) / np.std(test_ld)
test_fft = np.fft.fftshift(np.fft.fft(test_norm, 1024))
fig = plt.figure()
fig.set_size_inches(20,10)
plt.semilogy((np.abs(test_fft[513:])**2))
```

```python
fft_5 = [np.fft.fftshift(np.fft.fft(ld, 1024)) for ld in ldsnorm_5]
psd_5 = [np.abs(ft)**2 for ft in fft_5]
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))
for dd in psd_5:
    plt.semilogy(1.0 / faxis[513:],dd[513:])
    plt.xlim((0,40))
```

```python
psd_avg_5 = np.mean(psd_5, axis=0)
psd_norm_5 = psd_avg_5 / sum(psd_avg_5)
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))

plt.semilogy(1.0 / faxis[513:], psd_norm_5[513:])
plt.xlim((0,40))
```

```python
comb_20 = np.array([0 if (nn%36 != 0 and nn%36 != 1) else 1 for nn in range(714)])
len(comb_20)
```

```python
plt.plot(comb_20)
```

```python
comb_fft = np.fft.fftshift(np.fft.fft(comb_20, 1024))
comb_psd = np.abs(comb_fft[513:])**2
comb_psd /= np.sum(comb_psd) / len(comb_psd) #normalize
plt.plot(comb_psd)
```

```python
import scipy.signal
```

```python
gauss_env = scipy.signal.gaussian(714, 50 / PIXSIZE)
plt.plot(gauss_env)
```

```python
bfield = comb_20 * gauss_env
plt.plot(bfield)
```

```python
bf_fft = np.fft.fftshift(np.fft.fft(bfield, 1024))
bf_psd = np.abs(bf_fft)**2
bf_psd /= np.sum(bf_psd) #normalize
plt.plot(bf_psd)
```

```python
fig = plt.figure()
fig.set_size_inches(20,10)
faxis = np.fft.fftshift(np.fft.fftfreq(1024, PIXSIZE))

plt.semilogy(1.0 / faxis[513:], bf_psd[513:])
plt.semilogy(1.0 / faxis[513:], psd_norm_5[513:])

plt.xlim((0,40))
```
